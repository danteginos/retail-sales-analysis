### Calcula el total de ventas por categoría de producto.

```python
filtro = df.groupby(['Product Category'])["Total Amount"].sum()
print(filtro)
```

### Calcula el promedio de ventas diarias por categoría de producto.


```python
# se agrupa los datos por categoria y fecha y se aplica el promedio del total de ventas
df.groupby(["Product Category", "Date"])["Total Amount"].mean()
```

### Identifica las categorías de productos con mayores y menores ventas.

```python
# se agrupa los datos por Categoria con funcion groupby, se aplica la suma por el total de cada categoria, luego se ordenan los datos en orden descendente
# funcion sort_value expuesta en un ejemplo

fil_1 = df.groupby(["Product Category"])["Total Amount"].sum().sort_values(ascending=False)
print(fil_1)

```

4. Manipulación de Datos:

*  Filtra los datos para mostrar solo las ventas de una categoría de producto específica.

```python
categoria = "Clothing"
df_1 = df[df["Product Category"] == categoria]
fil_1 = df_1.groupby(["Product Category"])["Total Amount"].sum()
print(fil_1)
```

+ Realiza operaciones de suma, resta, multiplicación y división en los datos para obtener estadísticas adicionales.

```python
# Las  columnas deben contener numeros
"""
Se crea una columna nueva con la suma de las las columnas 'Total Amount' y 'Precio per Unit' y
se agrega al dataframe
"""

df["Suma Price per Unit y Total Amount"] = df["Price per Unit"] + df["Total Amount"]
print("\n",df[['Price per Unit', 'Total Amount', 'Suma Price per Unit y Total Amount']].head())

df["Resta Price per Unit a Total Amount"] = df["Total Amount"] - df["Price per Unit"]
print("\n",df[['Total Amount', 'Price per Unit', 'Resta Price per Unit a Total Amount']].head())

df["Multiplica Price per Unit por Total Amount"] = df["Price per Unit"] * df["Total Amount"]
print("\n",df[['Price per Unit', 'Total Amount', 'Multiplica Price per Unit por Total Amount']].head())

df["Divide Total Amount por Price per Unit"] = df["Total Amount"] / df["Price per Unit"]
print("\n",df[['Total Amount', 'Price per Unit', 'Divide Total Amount por Price per Unit']].head())

```
